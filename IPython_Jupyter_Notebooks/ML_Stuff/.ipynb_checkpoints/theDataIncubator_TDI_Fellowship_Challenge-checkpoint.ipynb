## Q6:
**Calculate the conditional probability of each violation category occuring given that it occured in a given council district. What is the median of these conditional probability? Do not consider violation category / council district pairs that had less than 50 counts.**

### Answer:
Conditional probability is defined as the likelihood of an event or outcome occurring, based on the occurrence of a previous event or outcome. Conditional probability is calculated by multiplying the probability of the preceding event by the updated probability of the succeeding, or conditional, event.

Because, the Conditional Probability Formula is P(B|A) = P(A and B) / P(A), following is how I proceed to find the answer:
* First calculate the number of violations in each category and a given council district (groupby on both of these columns?)
    * **Get P(A and B):** Get the ratio of violations in each category and in a given district w.r.t the total violations (for all categories and districts).
    * **Get P(B):** Get the ratio of the violations in a district w.r.t. the total violations (for all categories and districts)? Here use groupby only on 'districts'.
    * **Get P(A/B) = P(A and B)/P(B):** Then divide the first by the latter, which gives the conditional probability P(aCategory|aDistrict).
* Then find the median of all of the values of P(aCategory|aDistrict).

```python
display(dfpkl.columns)
#dfgpVD = dfpkl.groupby(["Violation_Category","CouncilDistrict"]).count()
dfgpVD = dfpkl.groupby(["Violation_Category","CouncilDistrict"])["Violation_Category"].count()
display(dfgpVD)
print(type(dfgpVD[["Dangerous Building"]]))#,["A"]])
print(dfgpVD[["Dangerous Building"]])#,["A"]])
dfgpVD = dfgpVD.to_frame().rename(columns={'Violation_Category':'Violations'}).reset_index()
display(dfgpVD)
print(dfgpVD["Violations"][0])
print(dfgpVD["Violations"][dfgpVD['Violation_Category']=='Nuisance'][dfgpVD['CouncilDistrict']=='K'])
```

## Q7:
**How big is Houston in terms of square miles? We can estimate the area by using the latitude and longitude of each of Houston's ZIP codes. These coordinates are in the population by ZIP code dataset. For each ZIP code, determine the distance to each ZIP code. With all of these distances, estimate the radius by usig this formula:**

    radius = 45*pi*D¯/128

**where D¯ is the average of all the calculated distances. Take the estimated radius and calculate the square miles of a circle with this radius. For calculating distances, please use the spherical Earth projected to a plane equation. Use the radius of the Earth as 6371 km. For conversion between kilometers and miles, use that one mile is equal to 1.60934 kilometers.**

### Answer:
Here is how I planned to go about before I did the actual calculations to get to the answer:
* First get average D 
     * Get the list of zip-codes (for which the coordinates 'lat' and 'long' are available). 
     * Out of this list and the corresponding latitudes and longitudes, make two copies of the same dataframes. Call one df1 and another df2 etc. 
          * Make sure the coordinates are in radians, rather than in degrees.
          * We can append new columns with radian values of latitudes and longitudes that we already have in degrees.
     * Use these dataframes in the `pairwise()` method of `sklearn.neighbors.DistanceMetric.get_metric('haversine')` to calculate the distances between each pair. This will give us a 'wide format' table of all possible distances between the zip-codes, which can then be converted into 'Long format' data set (called 'unpivoting') using `pd.melt(..)` method. From, this long format dataframe, it's a breeze to calculate the average D.
* Use this average D on the above given formula for effective 'radius' of the equivalent circular Texas.
* Use this radius on the formula for the area of a circle.


```python
#Get the zip codes and the latitudes and longitudes
#dfZipGp3 = dfZipGp2[dfZipGp2.Zip.isin(dfpop2['Zip Code'])]
#
#Make two copies of a  dataframe that has only the zip-code and the location coordinates
#dfpop.head() 
dfLocationsA = dfpop2[['Zip Code', 'Location']]
dfLocationsB = dfpop2[['Zip Code', 'Location']]
dfLocationsA.head()
```

```python
deg2rad = 0.01745329251 # pi/180
dfLocationsA[['latitude','longitude']] = dfLocationsA.Location.str.split(", ",expand=True)
dfLocationsB[['latitude','longitude']] = dfLocationsB.Location.str.split(", ",expand=True)
dfLocationsA = dfLocationsA.drop('Location', 1) #where 1 is the axis number (0 for rows and 1 for columns.)
dfLocationsB = dfLocationsB.drop('Location', 1)
dfLocationsA.head(2)
```

```python
dfLocationsA['latitude'] = pd.to_numeric(dfLocationsA['latitude'], downcast="float") #errors='coerce')
dfLocationsA['longitude'] = pd.to_numeric(dfLocationsA['longitude'], downcast="float") #errors='coerce')
dfLocationsB['latitude'] = pd.to_numeric(dfLocationsB['latitude'], downcast="float") #errors='coerce')
dfLocationsB['longitude'] = pd.to_numeric(dfLocationsB['longitude'], downcast="float") #errors='coerce')

dfLocationsA["latRadA"] = dfLocationsA["latitude"] * 0.01745329251 
dfLocationsA["longRadA"] = dfLocationsA["longitude"] * deg2rad
dfLocationsB["latRadB"] = dfLocationsB["latitude"] * 0.01745329251 
dfLocationsB["longRadB"] = dfLocationsB["longitude"] * deg2rad

display(dfLocationsA.head())
display(dfLocationsB.head())
```

```python
# Idea from: https://medium.com/@danalindquist/finding-the-distance-between-two-lists-of-geographic-coordinates-9ace7e43bb2f
import sklearn
from sklearn.neighbors import DistanceMetric

dist = sklearn.neighbors.DistanceMetric.get_metric('haversine')

distMatrix = (dist.pairwise
    (dfLocationsA[['latRadA','longRadA']],
     dfLocationsB[['latRadB','longRadB']])*3959
)
# Note that 3959 is the radius of the earth in miles

dfDistMatrix = (
    pd.DataFrame(distMatrix,index=dfLocationsA['Zip Code'], 
                 columns=dfLocationsB['Zip Code'])
)

display(dfDistMatrix.head())
display(dfDistMatrix.shape)
```

This table above displays the distance output (between all possible pairs of zip-codes in TX) in ‘wide format’. ‘Long format’ of this data is a list of each pair of cities and the distance between them. To do this conversion we unpivot the wide format table as follows:

```python
# Unpivot this dataframe from wide format to long format.
# When you unpivot, the data in the pivot table becomes a
# column named 'value'. Rename this column to 'miles' for clarity.
dfDistLong = (
    pd.melt(dfDistMatrix.reset_index(),id_vars='Zip Code')
)
dfDistLong = dfDistLong.rename(columns={'value':'miles'})

display(dfDistLong.head())
display(dfDistLong.shape)
```

```python
D_average = dfDistLong['miles'].mean()
D_average
```

```python
Radius = 45*np.pi*D_average/128

Area = np.pi * Radius**2

print(Radius, Area)
```

Initially, I planned to use the following functions of my own for the Haversine Formula. But, later I decided to use sklearn method(s):

```python
#Haversine Formula (from R.W. Sinnott, "Virtues of the Haversine",
#    Sky and Telescope, vol. 68, no. 2, 1984, p. 159):
#
#'lat' is latitude, 'lon' is longitude, R is earth’s radius (mean radius = 6,371km);
#note that angles need to be in radians to pass to trig functions!
#
# Distances are returned in Kilometers (km)
#
import math
def getDistanceHaversineDeg(lon1, lat1, lon2, lat2):
    deg2rad = 0.01745329251 # rad2deg = 57.2957795131
    dlon = (lon2 - lon1)*deg2rad
    dlat = (lat2 - lat1)*deg2rad
    R = 6371 #Earth's mean radius in km
    a = (math.sin(dlat/2))**2 + math.cos(lat1) * math.cos(lat2) * (math.sin(dlon/2))**2
    c = 2 * math.asin(min(1,math.sqrt(a)))
    d = R * c
    return d

def getDistanceHaversineRad(lon1, lat1, lon2, lat2):
    dlon = lon2 - lon1
    dlat = lat2 - lat1    
    R = 6371 #Earth's mean radius in km
    a = (math.sin(dlat/2))**2 + math.cos(lat1) * math.cos(lat2) * (math.sin(dlon/2))**2
    c = 2 * math.asin(min(1,math.sqrt(a)))
    d = R * c
    return d
```

# Challenge 2:
Pacman lives in a space that is effectively a torus - the left edge connect to the right, and the top edge connects to the bottom. We will set up an N by N grid in this space (so that 1 connects to N, using 1 based counting). We will measure distance ignoring this wrapping, so that N and 1 are N−1 spaces apart.
Without Pacman present, the ghosts randomly walk about this grid. Each ghost will pick one of the four cardinal directions (up, down, left, right) and move a number of spaces in that direction at each step.

### Krishna's Comments/Notes
All of these questions should be answerable in analytical/theoretical terms with some rigor of mathematical/statistical derivations. However, for this challenge, I will resort to mostly the Monte Carlo methods of conducting a fairly large number of random trials and counting various outcomes as needed.

<font color="green"> 
  Regarding the geometry, keep in mind the following key facts: <b>1 connects to N, using 1 based counting</b> and <b>we measure distances ignoring this wrapping, so that N and 1 are N−1 spaces apart.</b> which means, the curvature due to wrapping doesn't matter, just treat the distances just like in the ordinary plane.
    
To help myself on this geometry, I imagined a square sheet of some flexible material (say paper or rubber) with N by N grid, which is first rolled (edge-to-edge) into a right-circular-cylinder, then it is bent into a circle connecting the open ends of the cylinder, thus making a donut/doughnut/torus (this process of transformation would stretch the sheet that falls on the outer side and compress/contract/crumple that on the inner side so that it conforms to the differences in the inner and outer radii).
</font>

Because this geometry wraps, we should be mindful of this fact when considering the distances during a random walk. For example, in a 9-by-9 grid when a walk crosses the edge (i.e., wraps) along the x-direction (which is the line with x=0 where x=9 coincides), then incrementing x = 9 by 1 doesn't give x = 10, rather it gives x = 1. Similarly, if we're decrementing 0 by 1, it should be 8 instead of -1 and so on. The same fact applies to the y-coordinates too. This means, the values of the coordinates will always be a positive value ranging from 0 to N.

### Q11:
**We put Clyde on a 9 by 9 grid, starting at (5,5). Clyde is slow, and always takes steps of size 1. What is his expected distance from the starting position after 10 steps?**

```python
import random

#This works only when step size is always 1, doesn't work if >1
def applyWrappingCorrection(x,y, nGrid):
    """Apply corrections to the coordinates due to the wrapping of the geometry."""
    if x == nGrid + 1:
        x = 1
    if x == -1:
        x = nGrid - 1
    if y == nGrid + 1:
        y = 1
    if y == -1:
        y = nGrid - 1
        
    return x, y

# This new method is needed because there are cases when step size is 1 or 2 (Q13 ...)
#.     or when it is 1 or 2 or 3 (Question 16)
def applyWrappingCorrectionForAnyStepSize(x,y, nGrid):
    """Apply corrections to the coordinates due to the wrapping of the geometry."""
    if x - nGrid > 0:
        x = x - nGrid
    if x < 0:
        x = nGrid + x
    if y - nGrid > 0:
        y = y - nGrid
    if y < 0:
        y = nGrid + y
        
    return x, y


def randomWalk(n, X0, nGrid):
    """Return coordinates after 'n-steps' random walk. """ 
    x, y = X0[0], X0[1]
    for i in range(n):
        #dx = "differencce in x",    and dy = "difference in y"     
        (dx, dy) = random.choice([(0,1), (0,-1), (1,0), (-1,0)])
        x += dx
        y += dy
        #x, y = applyWrappingCorrection(x,y, nGrid)
        x,y = applyWrappingCorrectionForAnyStepSize(x,y, nGrid)
    
    
    return (x, y)


# For a test, let's take 25 random walks (it's not 25 steps of a single walk, hence n is not 25)
#.     On each walk, we'll take 10 steps
Xstart = [5,5]
nGrid = 9 # we have a 9-by-9 grid
for i in range(25):
    walk = randomWalk(10, Xstart, nGrid)
    print(walk, "Distance from home = ", 
          abs(walk[0] - Xstart[0]) + abs(walk[1] - Xstart[1]))
```

```python
# Now let's calculate an average distance for N random walks each with 10 steps
def getAverageRWDistance(numOfRandomWalks, numOfSteps, Xstart, nGrid):
    totalDistance = 0
    for i in range(numOfRandomWalks):
        walk = randomWalk(numOfSteps, Xstart, nGrid)
        #walk = random_walk_2(numOfSteps)
        totalDistance += abs(walk[0] - Xstart[0]) + abs(walk[1] - Xstart[1])
        
    return totalDistance/numOfRandomWalks
```

```python
grandAverage = 0
Xstart = [5, 5]
nGrid = 9
for i in range(20):
    average = getAverageRWDistance(100000, 10, Xstart, nGrid)
    grandAverage += average
    print("Average distance: ", average)
    
print("And the grand average is: ", grandAverage/20)
```

## Q13:
**We put Blinky on the same grid, but start him at (2,2). He is faster, and will takes steps of either size 1 or size 2 (with equal probability). After 10 steps, what is the probability that his walk "wraps" (crosses the edge and goes directly from N to 1 or vice versa) at least once?**

### Answer:

We should understand that the starting point is (2,2) and the edge is 7 steps away. And, not all 7 steps lead to the edge because the walk is not straight most of the time. However, we can be sure that if the final value (after 10 steps) of either of coordinates (x and y) is equal to or greater than 9, it wraps. These walks can be accumulated in order to evaluate the percentage and thus the probability of such "wraps". Now, during a single random walk, there could be more than 1 "wraps" or "crossings" because it may cross, come back in and cross again by turning back. Since the problem is about "wraps" at least once, we should make sure, we count such walks only once. So, inside the loop for a single walk, we should keep track of the number of crossings and then increment our global crossing counter only when the local crossing counter changes from 0 to 1. Once the loop for a single walk is over, we should reset the local wrap-counter to 0.

```python
import random

def randomWalkStepSize1or2(n, X0, printWrapFound):
    """Return coordinates after random walk with n-steps starting from X0. """ 
    x, y = X0[0], X0[1]
    wrappedAtLeastOnce = False
    for i in range(n):
        stepSize = random.choice([1,2])
        #dx is x-increment/decrement  and   dy is y-increment/decrement    
        #(dx, dy) = random.choice([(0,1), (0,-1), (1,0), (-1,0)])
        (dx, dy) = random.choice([(0,stepSize), (0,-stepSize), (stepSize,0), (-stepSize,0)])
        x += dx
        y += dy
        if x > nGrid or y > nGrid:    # 6 is = 10 - 2 - 1 = n - X0[0] - 1 etc.
            wrappedAtLeastOnce = True
            if printWrapFound == True:
                print("We have a 'wrap'")
                
        #x, y = applyWrappingCorrection(x,y, nGrid) # This doesn't work when step size can be > 1
        x, y = applyWrappingCorrectionForAnyStepSize(x,y, nGrid)
    
    return (x, y, wrappedAtLeastOnce)


# For a test, let's take 25 random walks (not 25 steps so n is not 25)
#.     On each walk, we'll take 10 steps
Xstart = 2, 2
for i in range(25):
    walk = randomWalkStepSize1or2(10, Xstart, True)
    print(walk, "Distance from home = ", 
          abs(walk[0] - Xstart[0]) + abs(walk[1] - Xstart[1]))
```

As we see above in the problem statement before Q11, we will measure distance ignoring the wrapping, so that N and 1 are N−1 spaces apart, i.e., we'll measure the distance only going along one direction along each of the two axes (Otherwise, we could say N and 1 are just 1 space apart (because N meets 0, right at the edge, by going around).

Now, if the distance of the random walk is equal to or more than 7 spaces/steps/units (i.e., more than 6 units or if the condition **(x > 6 or y > 6)** is satisfied), then it wraps.

So, let's do a large # of loops and count how many of them have at least one wrap and that will give us the probability

```python
# Check if there is ever a wrap (looping it a larger # of times):
printWrapFound = False
nWalksWithAtLeastOneWrap = 0
nLoops = 100000

for i in range(nLoops):
    walk = randomWalkStepSize1or2(10, Xstart, printWrapFound)
    if walk[2] == True:
        nWalksWithAtLeastOneWrap += 1
        
print("Probability of Blinky's walks that wrap at least once", float(nWalksWithAtLeastOneWrap)/nLoops)
```

## Q14:
**In Blinky's random walk, some of his paths will wrap (cross the edge) at least once and some will not (so zero edge crossings). What is the difference between his average final distance from his starting position for those walks that involved wrapping and his average final distance for those walks that did not?**

#### Answer:
I will Use two counters to count the walks of the two categories and get the difference between the corresponding distances

```python
# Now let's calculate an average distance for N random walks each with 10 steps
def getAverageRWDistance(numOfRandomWalks, numOfSteps, Xstart):
    totalDistanceWithWraps = 0
    totalDistanceWithoutWraps = 0
    walkCounts = [0, 0]
    printWrapFound = False
    for i in range(numOfRandomWalks):
        walk = randomWalkStepSize1or2(10, Xstart, printWrapFound)
        totalDistance = abs(walk[0] - Xstart[0]) + abs(walk[1] - Xstart[1])
        if walk[2] == True:
            totalDistanceWithWraps += totalDistance
            walkCounts[0] += 1
        else:
            totalDistanceWithoutWraps += totalDistance
            walkCounts[1] += 1
        
    return [float(totalDistanceWithWraps)/walkCounts[0], 
            float(totalDistanceWithoutWraps)/walkCounts[1]]
```

```python
import pandas as pd
df = pd.DataFrame(columns=['AvgDistWithWraps', 'AvgDistWithoutWraps'])

numOfRandomWalks, numOfSteps = 100000, 10
Xstart = [2, 2]
for i in range(10):
    dists = getAverageRWDistance(numOfRandomWalks, numOfSteps, Xstart)
    df = df.append({'AvgDistWithWraps': dists[0], 'AvgDistWithoutWraps': dists[1]}, 
                   ignore_index = True)
    print(dists)

print('The Mean values of the two distances are: ', 
      df['AvgDistWithWraps'].mean(), df['AvgDistWithoutWraps'].mean())
print('And, the difference between the two is: ',
      df['AvgDistWithWraps'].mean() - df['AvgDistWithoutWraps'].mean())
```

```python
# Now let's calculate an average distance for N random walks each with 'randomWalkSize' (eg. 10, 25, 50 or 100 steps)
#.    and stepOptions == 123 which means "steps of either size 1, 2, or 3 (with equal probability)"

def getAverageRWDistanceSteps1or2or3(numOfRandomWalks, randomWalkSize, Xstart, printWRapFound, stepOptions, nGrid):
    """Return average distance of a random walk with given value of randomWalkSize, 
    Xstart, printWRapFound, stepOptions """ 
    totalFinalDistanceForAllWalks = 0
    for i in range(numOfRandomWalks):
        walk = randomWalkStepSize1or2or3(randomWalkSize, Xstart, printWrapFound, stepOptions, nGrid) #12)
        finalDistanceOfCurrentWalk = abs(walk[0] - Xstart[0]) + abs(walk[1] - Xstart[1])
        totalFinalDistanceForAllWalks += finalDistanceOfCurrentWalk
    
    averageFinalDistance = float(totalFinalDistanceForAllWalks)/numOfRandomWalks
    print('average distance with ', randomWalkSize, ' steps = ', averageFinalDistance)
    return averageFinalDistance

```
