I wasn't sure how to deal with space separated data file where there could be different number of white spaces between datas from different columns. So, I googled about it and landed on this forum () which said the following:

From the [documentation](http://pandas.pydata.org/pandas-docs/stable/io.html), you can use either a regex or `delim_whitespace`:

```py
>>> import pandas as pd
>>> for line in open("whitespace.csv"):
...     print repr(line)
...     
'a\t  b\tc 1 2\n'
'd\t  e\tf 3 4\n'
>>> pd.read_csv("whitespace.csv", header=None, delimiter=r"\s+")
   0  1  2  3  4
0  a  b  c  1  2
1  d  e  f  3  4
>>> pd.read_csv("whitespace.csv", header=None, delim_whitespace=True)
   0  1  2  3  4
0  a  b  c  1  2
1  d  e  f  3  4
```
In that documentation, I see the following in addtion to many other info:

**delim_whitespaceboolean, default False**

    Specifies whether or not whitespace (e.g. ' ' or '\t') will be used as the delimiter. Equivalent to setting sep='\s+'. If this option is set to True, nothing should be passed in for the delimiter parameter.


````python
dfs4 = pd.read_csv('kpSpaceSeparatedData.txt', delimiter='\s+')
dfs4.head()
````

````python
dfs5 = pd.read_csv('kpSpaceSeparatedData.txt', delim_whitespace=True)
dfs5.head()
````

#### Adding a new column in an alternative way:

````python
# The ':' before comma in 'iloc' means include all the rows
# The 4 in 4:10 is for the fourth column (first numerical column)
# The 10 is for (n+1), where n is the last numerical column that we want to add together
#  axis=1 means adding horizontally,  axis=0 means adding vertically
df['Total'] = df.iloc[:, 4:10].sum(axis=0)
df.head(3)
````

````python
# The ':' before comma in 'iloc' means include all the rows
# The 4 in 4:10 is for the fourth column (first numerical column)
# The 10 is for (n+1), where n is the last numerical column that we want to add together
#  axis=1 means adding horizontally,  axis=0 means adding vertically
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(3)
````

````python
# Selecting all rows whose 'Name' contains 'Mega'
df.loc[df['Name'].str.contains('Mega')].head()
````

````python
# Reverse of above one - in other words, selecting all rows whose 'Name' doesn't have 'Mega'
# Instead of !, we use ~ (squiggly line?)
df.loc[~df['Name'].str.contains('Mega')].head()
````

````python
# And taking this even a step farther, the contains() function is very very powerful because
#   in addition to just doing exact words, we can also pass in RegEx expressions and do all
# sorts of complicated filtering with this
# Let's look at one example:

# 'Type 1' either 'Grass' or 'Fire'
# First import 'regular expressions'
import re

#df.loc[df['Name'].str.contains('Mega')].head()
#df.loc[df['Type 1'].str.contains('fire|grass', regex=True)].head() #Only gave headers row due to missing capitalization
#df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)].head(7) #Works
df.loc[df['Type 1'].str.contains('Fire|Grass', flags=re.I, regex=True)].head(6)  #Equivalent: ignores capitalization
````

````python
# Selecting those with 'Pi' or 'pi'
# Below in pi[a-z]*, it starts with 'pi' followed by any letter [a-z], with * indicating 'one or more'
df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)].head(3)
````

````python
# Following is for selecting those which starts with Pi or pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)].head(3)
````

## Aggregate Statistics (Groupby)


````python
#df = pd.read_csv('modified.csv')

df.groupby(['Type 1']).mean()
#kp: Here it seems we're getting average of all the numerical data in each group, where a group contains 
#    all the rows that has the same value of 'Type 1'. For example, one group contains those rows with
#.   Type 1 = Bug and the corresponding averages/means of HP, Attack etc can be found in the first row below.
````

We can also sort above results

````python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
````

````python
gSum = df.groupby(['Type 1']).sum()
gSum
````

````python
df.groupby(['Type 1']).count()
````

Above we see that the count is mostly the same for many columns but different/less for some columns. That's because some data is non-numerical type and hence that row is not counted. In order to count all rows for each group, we can introduce a new column called "count" with a value of 1 for each row, then we can re-execute the groupby().count() as follows:

````python
df['count'] = 1 #kp adding a new column for 'count' with a value of 1 for each row
df.head()
````

````python
gCounts = df.groupby(['Type 1']).count()
print(type(gCounts))
gCounts
````

The two steps (adding 'count' column and counting) can be done simultaneously as follows. Before that let me first drop the newly added 'count' column.

````python
#df = df.drop(columns=['count'])
#df.head(3)
````

````python
#df.groupby(['Type 1']).count()

#Here, the above line gives counts of all groups from all columns, but
#      the following line gives counts of all groups from only 'count' column.

df.groupby(['Type 1']).count()['count'] #Makes life easier by doing both steps all at once.
````

**We can also group by multiple parameters (for counts of subsets/subgroups within a group)**

````python
df.groupby(['Type 1', 'Type 2']).count()['count']
#kp: In the result, we'll see that (Bug, Electric) comes 2 times, (Water,Ice) comes 3 times etc.
````

## Working with large amounts of data

Pandas allows reading/working with chunks of big sized data files (let's say 20 GB files).

````python
new_df = pd.DataFrame(columns=df.columns) #An empty dataframe with the same columns as df.

#for df in pd.read_csv('modified.csv', chunksize=5): #kp: chunksize=5 means 5 rows are passed to reader at a time
for df in pd.read_csv('pokemon_data.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results]) #This concats or appends the dataframes (results are dataframes too).
    
#new_df
````
