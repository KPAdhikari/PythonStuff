<a id='scilab_kernel'></a>
## Installing Scilab Kernel
Reference: https://github.com/calysto/scilab_kernel

First of all,  I installed (7/28/17) Scilab 6.0.0 (Scilab is one of the open-source alternatives to MATLAB, another popular alternative is [GNU Octave](https://www.gnu.org/software/octave/)) by dowloading the [scilab-6.0.0-x86_64.dmg](http://www.scilab.org/download/6.0.0/scilab-6.0.0-x86_64.dmg) file (222 MB), opening it and dragging Scilab icon to the Applications folder (from the new window that dmg file opened). Now it shows up in the Launcher. (References: http://www.scilab.org/)

Then, I used the following commands to install it after Scilab was installed:
```
pip install scilab_kernel
python -m scilab_kernel.install
```
